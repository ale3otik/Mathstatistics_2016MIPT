```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
num_tests = 10
N = 10**4
theta = 1

OX = [i for i in range(1,N+1)]
colors = ['b','r','g','c','y',"orange","#821e3c","#177245","k--","g--"]
description = ["k = " + str(i) for i in range(1,num_tests+1)]
               
def calc_evaluation_quality(result , k) :
    sum = 0
    fact = 1;
    for i in range(k) :
        fact *= i+1

    for i in range(N) :
        sum += X[i] ** k
        
        d = sum / (i+1)
        result[i] = abs((fact / d)**(1/k) - theta)


#sample generation
for n in range(1,N+1) :        
    X = stats.expon.rvs(scale = 1/theta, size = N)
        

result = [np.zeros(N) for i in range(num_tests)]
s = 0
for k in range(1,num_tests+1):
    calc_evaluation_quality(result[k-1] , k)
    

plt.title(r'$|\theta_{k}(n) - \theta|$')
plt.xlabel(r'$n$')
for i in range(num_tests): 
    plt.plot(OX,result[i],colors[i],label = description[i])

plt.legend(loc=(1.05,0.1))
plt.axis([0, N, -0.01, 0.3])  
plt.show()


plt.title(r'$|\theta_{k}(n) - \theta|$')
plt.xlabel(r'$n$')
for i in range(num_tests): 
    plt.plot(OX,result[i],colors[i],label = description[i])

plt.legend(loc=(1.05,0.08))
plt.axis([0, N, -0.001, 0.12])  
plt.show()

OX = [k for k in range(1,num_tests + 1)]
plt.title(r'$|\theta_{k}(N) - \theta|$')
plt.xlabel(r'$k$')
ndiff = [result[i][N-1] for i in range(num_tests)]
plt.plot(OX,ndiff,'.-r')
plt.show()

#:
#1)        ""    ,
#         k
#2)    = 1    \theta
```
