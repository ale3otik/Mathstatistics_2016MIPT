```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import math
import random
%matplotlib inline
```

```python
N = 1000
K = 10000
sample = sps.norm.rvs(size = N*K)
sample = np.array(sample).reshape(K,N) # X transp
```

```python
narange = np.arange(0,N,1)
likly = [[np.mean(sample[j][:n+1]) 
                    for j in range(K)]
             for n in narange]
theta2 = [[(np.max(sample[j][:n+1]) + np.min(sample[j][:n+1])) * (0.5) 
                         for j in range(K)]
                    for n in narange]
```

```python
sigm_lh = [np.var(likly[n]) for n in narange]
sigm_m2 = [np.var(theta2[n]) for n in narange]
plt.plot(narange , sigm_lh)
plt.title('likelyhood')
plt.ylim(0,0.2)
plt.show()
plt.plot(narange , sigm_m2)
plt.title('1/2 (min + max)')
plt.ylim(0,0.2)
plt.show()
```

Оценка максимального правдоподобия лучше, так как имеет меньшую дисперсию
